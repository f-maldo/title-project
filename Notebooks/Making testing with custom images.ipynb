```python
%matplotlib inline
from keras import models
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from skimage import io
from skimage.transform import resize
```

```python
from scipy.io import loadmat
data = loadmat("../CompCars/data/misc/make_model_name.mat")
```

```python
model = models.load_model("../Models/mark2.h5")
```

```python
im = io.imread("../SingleTestings/toyota_corolla.jpg")
im = resize(im, (256, 256))
im = im.reshape((1, 256, 256, 3))
io.imshow(im[0]);
prediction = model.predict(im)
```

```python
predicted_class = np.argmax(prediction, axis=1)
print(predicted_class)
print("Predicted make and model: " + str(data["model_names"][predicted_class[0]][0]))
```
